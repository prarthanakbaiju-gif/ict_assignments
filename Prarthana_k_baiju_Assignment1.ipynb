4⃣ .Maximum Finder
Task:
● Create a 4×4 random matrix.
● Find the maximum value.
● Find its position (row, column).

```python
import numpy as np

A = np.random.randint(1, 10, (4, 4))

max_value = A.max()
position = np.unravel_index(A.argmax(), A.shape)

print(A)
print("Maximum value:", max_value)
print("Position:", position)
```

6. Simple Matrix Operations
Task:
● Create two 3×3 random matrices.
● Add them.
● Multiply them element-wise.
● Multiply them using matrix multiplication.

```python
import numpy as np

A = np.random.randint(1, 10, (3, 3))
B = np.random.randint(1, 10, (3, 3))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

print("Addition:\n", A + B)

print("Element-wise Multiplication:\n", A * B)

print("Matrix Multiplication:\n", A @ B)
```

8⃣ Filter Rows Based on Condition
Task:
● Create a 5×3 random matrix (values between 0–20).
● Print only rows where row sum > 30.

```python
import numpy as np

A = np.random.randint(0, 21, (5, 3))

print("Matrix:\n", A)

for row in A:
    if row.sum() > 30:
        print(row)
```

10.Simple Normalization
Task:
● Create a 3×3 random matrix.
● Divide each row by its row sum.

```python
import numpy as np

A = np.random.randint(1, 10, (3, 3))

row_sum = A.sum(axis=1, keepdims=True)

normalized = A / row_sum

print(normalized)
```

3. Create a 5×5 matrix filled with zeros except the border filled with 1.

```python
import numpy as np

A = np.zeros((5, 5), dtype=int)

A[0, :] = 1
A[-1, :] = 1
A[:, 0] = 1
A[:, -1] = 1

print(A)
```
