= (°C × 9/5) + 32.

3. Salary Slip
Given the basic salary of an employee, calculate the total salary including HRA (20%) and
DA (10%).
Sample Input:
Enter basic salary: 20000
Sample Output:
Net Salary: 26000.0

```python
s=int(input("Enter the basic salary "))
hr=s*20/100
da=s*10/100
m=s+hr+da
print('The net salary =',m)
```

4. Simple Interest
Write a program to calculate simple interest using the formula SI = (P × R × T) / 100.
Sample Input:
Principal: 10000
Rate of Interest: 5
Time (in years): 2
Sample Output:
Simple Interest = 1000.0

```python
p=int(input("Enter the principal amount : "))
r=int(input("Enter the rate of interest : "))
t=int(input('Enter the time : '))
si=(p*r*t)/100
print("simple interest : ",si)
```

6. BMI Calculator
Write a program to calculate the Body Mass Index (BMI) using the formula: BMI
weight(kg) / (height(m) × height(m))
Sample Input:
Weight: 60
Height: 1.65
Sample Output:
BMI = 22.04

```python
w=int(input('Enter the weight : '))
h=float(input('Enter the height : '))
bmi=w//(h*h)
print('BMi : ',bmi)
```

8. Area of a Triangle (Using Heron’s Formula)
A land surveyor wants to calculate the area of an irregular triangular plot.
Write a Python program to find the area of a triangle when all three sides are known,
using Heron’s formula:
�
� =𝑎+𝑏+𝑐
 2
 Area = √𝑠(𝑠 −𝑎)(𝑠−𝑏)(𝑠−𝑐)
 Sample Input:
Enter side a: 5
Enter side b: 6
Enter side c: 7
Sample Output:
Area of the triangle: 14.70

```python
import math
a=int(input('Enter the side a : '))
b=int(input('Enter the side b : '))
c=int(input('Enter the side c : '))
s=(a+b+c)/2
area=math.sqrt(s*(s-a)*(s-b)*(s-c))
print('Area of the triangle : ',area)
```

10. Electricity Bill Calculator
An electricity board charges consumers based on unit consumption. The rate per unit is
₹5.2. Additionally, there is a fixed charge of ₹100 on every bill. Write a Python program
that takes the number of units consumed and calculates the total bill amount.
Sample Input:
Enter number of units consumed: 250
Sample Output:
Electricity Bill: ₹1400.0

```python
c=int(input('Enter the number of units consumed : '))
r=5.2
f=100
t=c*r+f
print('Electricity Bill : ',t)
```
