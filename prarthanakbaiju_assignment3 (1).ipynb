#**2**. Write a Python function to sum all the numbers in a list.
Sample List : [18, 12, 30,10, 7]
Expected Output : **77**

```python
def sum_list(items):
  s = 0
  for x in items:
    s+=x
  return s
```

```python
list=[18,12,30,10,7]
print('sum is',sum_list(list))
```
